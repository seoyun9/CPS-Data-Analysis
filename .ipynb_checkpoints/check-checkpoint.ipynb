```python
import pandas as pd
import os
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
# Kaggle API 인증
api = KaggleApi()
api.authenticate()

# 데이터셋 다운로드
dataset_name = "stephanmatzka/predictive-maintenance-dataset-ai4i-2020"
api.dataset_download_files(dataset_name, path="./", unzip=True)

# CSV 파일 읽기 (파일명 확인 필요)
data = pd.read_csv("ai4i2020.csv", encoding="utf-8")
data = data.drop(columns=["Product ID", "UDI", "Type"])
data
```

```python
# 고장 유형
failure_cols = ["TWF", "HDF", "PWF", "OSF", "RNF"]

# 각 고장 유형별 데이터프레임
twf_df = data[data["TWF"] == 1]
hdf_df = data[data["HDF"] == 1] 
pwf_df = data[data["PWF"] == 1]
osf_df = data[data["OSF"] == 1]
rnf_df = data[data["RNF"] == 1]
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt


FEATURES = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]",
]


def plot_features(df, title_prefix=""):
    # 주어진 데이터프레임에서 5개 feature를 히스토그램으로 시각화
    plt.figure(figsize=(15, 8))

    for i, col in enumerate(FEATURES, start=1):
        if col not in df.columns:
            continue
        plt.subplot(2, 3, i)
        df[col].hist(bins=30, edgecolor="black")
        plt.title(col)
        plt.xlabel(col)
        plt.ylabel("Count")

    plt.suptitle(f"{title_prefix} Feature Distributions", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


plot_features(twf_df, title_prefix="TWF")
```

```python
plot_features(hdf_df, title_prefix="HDF")
```

```python
plot_features(pwf_df, title_prefix="PWF = 1")
```

```python
plot_features(osf_df, title_prefix="OSF = 1")
```

```python
plot_features(rnf_df, title_prefix="RNF = 1")
```

```python
import seaborn as sns

# 1) Machine failure 기준 Boxplot

def plot_box_by_failure(df):
    plt.figure(figsize=(15, 8))
    for i, col in enumerate(FEATURES, start=1):
        plt.subplot(2, 3, i)
        sns.boxplot(data=df, x="Machine failure", y=col)
        plt.title(f"{col} vs Machine failure")
    plt.tight_layout()
    plt.show()

# 예시 실행
# plot_box_by_failure(data)
```

```python
# 2) Rotational speed vs Torque 산점도 (고장 여부 색 구분)

def plot_speed_torque_scatter(df):
    plt.figure(figsize=(7, 6))
    sns.scatterplot(
        data=df,
        x="Rotational speed [rpm]",
        y="Torque [Nm]",
        hue="Machine failure",
        alpha=0.5,
    )
    plt.title("Rotational speed vs Torque (colored by Machine failure)")
    plt.show()

# 예시 실행
# plot_speed_torque_scatter(data)
```

```python
# 3) Tool wear vs Machine failure (분포 비교)

def plot_tool_wear_by_failure(df):
    plt.figure(figsize=(7, 6))
    sns.histplot(
        data=df,
        x="Tool wear [min]",
        hue="Machine failure",
        multiple="stack",
        bins=30,
    )
    plt.title("Tool wear distribution by Machine failure")
    plt.show()

# 예시 실행
# plot_tool_wear_by_failure(data)
```

```python
# 4) 고장 유형별 개수 / 비율 Barplot

def plot_failure_type_counts(df):
    failure_counts = df[["TWF", "HDF", "PWF", "OSF", "RNF"]].sum()
    plt.figure(figsize=(6, 4))
    sns.barplot(x=failure_counts.index, y=failure_counts.values)
    plt.title("Failure type counts")
    plt.ylabel("Count")
    plt.show()

# 예시 실행
# plot_failure_type_counts(data)
```

```python
# 5) 상관관계 히트맵

def plot_correlation_heatmap(df):
    plt.figure(figsize=(10, 8))
    sns.heatmap(df.corr(), annot=False, cmap="coolwarm", center=0)
    plt.title("Correlation Heatmap")
    plt.show()

# 예시 실행
# plot_correlation_heatmap(data)
```

```python
# 한 번에 주요 시각화를 모두 보고 싶을 때 쓸 수 있는 편의 함수

def run_basic_eda(df):
    plot_features(df, title_prefix="All data")
    plot_box_by_failure(df)
    plot_speed_torque_scatter(df)
    plot_tool_wear_by_failure(df)
    plot_failure_type_counts(df)
    plot_correlation_heatmap(df)

# 예시 실행
# run_basic_eda(data)
```
